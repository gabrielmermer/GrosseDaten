```python
import pandas as pd

# Read the CSV file
student_df = pd.read_csv("student-lpor.csv")

# View the first 5 rows
student_df
```

```python
# Remove duplicate records
student_df.drop_duplicates(inplace=True)

# Identify and handle missing values
student_df.dropna

# Display information about the cleaned DataFrame
student_df
```

```python
# prompt: visualize the age

import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(student_df['age'], bins=20)
plt.title('Distribution of Student Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# prompt: visualize the gender

plt.figure(figsize=(8, 8))
student_df['sex'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Student Gender')
plt.ylabel('')
plt.show()
```

```python
# Visualize the distribution of Dalc (daily alcohol consumption)
plt.figure(figsize=(8, 6))
sns.countplot(x='Dalc', data=student_df, palette='viridis')
plt.title('Alcohol consumption on the work day (categorical: from 1 - very low to 5 - very high)')
plt.xlabel('Alcohol consumption on the work day')
plt.ylabel('Count')
plt.show()
```

```python
# prompt: compare g1 grades with dalc

# Visualize the relationship between G1 and Dalc
plt.figure(figsize=(12, 6))
sns.boxplot(x='Dalc', y='G1', data=student_df, palette='plasma')
plt.title('Relationship between G1 Grades and Alcohol Consumption on Workdays')
plt.xlabel(' Alcohol Consumption (1: Very Low, 5: Very High)')
plt.ylabel('G1 Grade')
plt.show()
```

```python
# prompt: how many datapoints per dalc level

# Calculate the number of datapoints per Dalc level
dalc_counts = student_df['Dalc'].value_counts().sort_index()

print("Number of datapoints per Dalc level:")
dalc_counts
```

```python
# prompt: visualize walc

# Visualize the distribution of Walc (weekly alcohol consumption)
plt.figure(figsize=(8, 6))
sns.countplot(x='Walc', data=student_df, palette='viridis')
plt.title('Alcohol consumption on the weekend (categorical: from 1 - very low to 5 - very high)')
plt.xlabel('Weekend Alcohol Consumption (1: Very Low, 5: Very High)')
plt.ylabel('Count')
plt.show()

# Calculate the number of datapoints per Walc level
walc_counts = student_df['Walc'].value_counts().sort_index()

print("Number of datapoints per Walc level:")
walc_counts
```

```python
# prompt: relation with g1 and walc give the numbers of the median

# Visualize the relationship between G1 and Walc
plt.figure(figsize=(12, 6))
sns.boxplot(x='Walc', y='G1', data=student_df, palette='plasma')
plt.title('Relationship between G1 Grades and Alcohol Consumption on Weekends')
plt.xlabel('Weekend Alcohol Consumption (1: Very Low, 5: Very High)')
plt.ylabel('G1 Grade')
plt.show()

# Calculate the median G1 for each Walc level
median_g1_by_walc = student_df.groupby('Walc')['G1'].median()

print("Median G1 grade for each Walc level:")
median_g1_by_walc
```

```python
# prompt: correlation matrix with g1 and all other variables

# Select only the numerical columns
numerical_df = student_df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation of 'G1' with all other numerical columns
correlation_g1 = numerical_df.corr()['G1']

# Print the correlations
print("Correlation of 'G1' with other numerical variables:")
correlation_g1
```

```python
# prompt: take the average of fedu and medu and correlate it to G1, make it a bar chart

# Calculate the average of 'Fedu' and 'Medu'
student_df['avg_parental_edu'] = (student_df['Fedu'] + student_df['Medu']) / 2

# Calculate the correlation between the average parental education and G1
correlation_avg_edu_g1 = student_df['avg_parental_edu'].corr(student_df['G1'])

print(f"Correlation between Average Parental Education and G1: {correlation_avg_edu_g1:.2f}")

# Create a bar chart to visualize the average G1 for different average parental education levels
# Group by average parental education and calculate the mean of G1
avg_g1_by_avg_edu = student_df.groupby('avg_parental_edu')['G1'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='avg_parental_edu', y='G1', data=avg_g1_by_avg_edu, palette='viridis')
plt.title('Average G1 Grade by Average Parental Education Level')
plt.xlabel('Average Parental Education Level')
plt.ylabel('Average G1 Grade')
plt.xticks(rotation=0)
plt.show()
```

```python
# prompt: bar chart of study time and grade

# Assuming 'student_df' is your DataFrame and it has 'studytime' and 'G1' columns
plt.figure(figsize=(10, 6))
sns.barplot(x='studytime', y='G1', data=student_df, palette='viridis')
plt.title('Average G1 Grade by Study Time')
plt.xlabel('Study Time (1: <2h, 2: 2-5h, 3: 5-10h, 4: >10h)')
plt.ylabel('Average G1 Grade')
plt.show()
```

```python
# prompt: average alcohol consumption (walc and dalc) compared to age (bar chart)

# Group by age and calculate the mean of Dalc and Walc
alcohol_consumption_by_age = student_df.groupby('age')[['Dalc', 'Walc']].mean().reset_index()

# Melt the DataFrame to have 'age', 'alcohol_type', and 'average_consumption' columns
alcohol_consumption_melted = alcohol_consumption_by_age.melt(id_vars='age', var_name='alcohol_type', value_name='average_consumption')

plt.figure(figsize=(12, 7))
sns.barplot(x='age', y='average_consumption', hue='alcohol_type', data=alcohol_consumption_melted, palette='coolwarm')
plt.title('Average Daily and Weekly Alcohol Consumption by Age')
plt.xlabel('Age')
plt.ylabel('Average Alcohol Consumption')
plt.legend(title='Alcohol Type')
plt.show()
```

```python
# prompt: how many datapoints for 22 year olds

age_22_datapoints = student_df[student_df['age'] == 22].shape[0]
print(f"Number of datapoints for 22 year olds: {age_22_datapoints}")
```

```python
# prompt: relationship with absences, g1 with trend line

plt.figure(figsize=(10, 6))
sns.regplot(x='absences', y='G1', data=student_df)
plt.title('Relationship between Absences and G1 with Trend Line')
plt.xlabel('Number of Absences')
plt.ylabel('G1 Grade')
plt.show()
```
